```python
#make a function so if i call it it'll make a new model, like if i want KNN, or if I want SVC, trees, Naive Bayes
def model_maker(model):
    if model == "KNN":
        clf = KNeighborsClassifier(n_neighbors=10)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)

        # Calculate the confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        print("Confusion Matrix:\n", cm)

        print("made KNN")
    elif model == "NaiveBayes":
        clf = GaussianNB()
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)

        # Calculate the confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        print("Confusion Matrix:\n", cm)

        print("made NaiveBayes")
    elif model == "SVC":
        clf = SVC(random_state=42)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)

        # Calculate the confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        print("Confusion Matrix:\n", cm)

        print("made SVC")
    elif model == "RandomForest":
        clf = RandomForestClassifier(n_estimators=200, random_state=42)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)

        # Calculate the confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        print("Confusion Matrix:\n", cm)

        print("made Random Forest")
    else:
        raise ValueError("Unknown model name")
    return clf
    
        
```

```python
def fit_and_confmat(clf, X_train, y_train, X_test, y_test):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:\n", cm)
    return cm
```

```python
def accuracy_from_cm(cm):
    correct = np.trace(cm)          # sum of diagonal
    total = cm.sum()
    acc_pct = 100 * correct / total
    print("below are accurate values")
    print(correct)
    print("below are inaccurate values")
    print(total - correct)
    print(acc_pct)
    return acc_pct
```

```python
for name in ["KNN", "NaiveBayes", "SVC", "RandomForest"]:
#for name in ["RandomForest","KNN", "NaiveBayes", "SVC"]:
    clf = model_maker(name)
    cm = fit_and_confmat(clf, X_train, y_train, X_test, y_test)
    accuracy_from_cm(cm)                         # <-- use THIS model’s cm
    # (optional) compare with sklearn’s built-in accuracy:
    print("sklearn accuracy:", accuracy_score(y_test, clf.predict(X_test))*100)
```
